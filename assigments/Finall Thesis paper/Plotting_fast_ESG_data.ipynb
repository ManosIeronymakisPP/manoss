# Fast data analysis and plots 

```python
# Function to compute mean values for specific columns
def compute_means(df, columns):
    return df[columns].mean()

# Define the column ranges for each set
years = list(range(2005, 2025))
years_S = [f"{year}S" for year in years]
years_F = [f"{year}F" for year in years]

# Compute mean values
mean_ESG_S = compute_means(ESG, years_S)
mean_ESG_F = compute_means(ESG, years_F)
mean_ESGC_F = compute_means(ESGC, years_F)
mean_ESGControv_F = compute_means(ESGControv, years_F)
mean_ESGEP = compute_means(ESGEP, years_F)
mean_ESGGP = compute_means(ESGGP, years_F)
mean_ESGSP = compute_means(ESGSP, years_F)
```

```python
# Plotting
plt.figure(figsize=(12, 6))

# plt.plot(years, mean_ESG_S, label='Slow ESG scores ', marker='o', color='#004976')
plt.plot(years, mean_ESG_F, label='ESG scores', marker='o', color='#0097A9')
plt.plot(years, mean_ESGC_F, label='ESG Combined scores', marker='o', color='#004976')
plt.plot(years, mean_ESGControv_F, label='ESG Controversies scores', marker='o', color='#FF6A13')

plt.xlabel('Year')
plt.ylabel('Mean ESG,ESG Combined and ESG Controversies Scores')
plt.title('Comparison of mean ESG,ESG Combined and ESG Controversies scores for S&P500 companies')
plt.legend()
plt.grid(True)
plt.xticks(years)  # Set x-axis to show each year
plt.show()
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(years, mean_ESGEP, label='Environmental Pillar', marker='o', color='#004976')
plt.plot(years, mean_ESGGP, label='Governance Pillar', marker='o', color='#0097A9')
plt.plot(years, mean_ESGSP, label='Social Pillar', marker='o', color='#FF6A13')

plt.xlabel('Year')
plt.ylabel('Mean EnvironmentalPillar,GovernancePillar and SocialPillar')
plt.title('Comparison of mean  EnvironmentalPillar,GovernancePillar and SocialPillar for S&P500 companies')
plt.legend()
plt.grid(True)
plt.xticks(years)  # Set x-axis to show each year
plt.show()
```
